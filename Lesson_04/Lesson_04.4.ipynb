```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
df_cred = pd.read_csv('creditcard.csv')
df_cred.describe()
```

```python
plt.style.use('fivethirtyeight')
```

```python
df_cred_class = df_cred['Class'].value_counts()
df_cred_class
```

```python
df_cred_class.plot(kind = 'bar')
plt.show()
```

```python
df_cred_class.plot(kind = 'bar')
plt.yscale(value = 'log')
plt.show()
```

```python
plt.hist(df_cred.loc[(df_cred['Class'] == 1),'V1'], bins=20, ec = 'black', density = 1, alpha = 0.5, label="Class = 1", color = 'grey')
plt.hist(df_cred.loc[(df_cred['Class'] == 0),'V1'], bins=20, ec = 'black', density = 1, alpha = 0.5, label="Class = 0", color = 'red')
plt.legend(loc="upper left", frameon=False)
plt.xlabel("Class")
plt.rcParams.update({'font.size': 15})
plt.rcParams['figure.figsize']=7,9
plt.show()
```
