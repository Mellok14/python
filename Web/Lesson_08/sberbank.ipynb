```python
df_sber = pd.read_csv(r"C:\Users\Mell\PycharmProjects\Web\Lesson_08\opendata.csv", encoding = 'ANSI', sep=',')
df_sber.head(10)
```

Выведем типы данных и доступные регионы

```python
print(df_sber['name'].value_counts())
```

```python
print(df_sber['region'].value_counts())
```

```python
using_name = 'В среднем руб. на текущем счете на человека'
using_region = 'Мурманская область'
date_start = '2016-01-15'
date_end = '2019-01-15'
```

```python
df_sber_print = df_sber.loc[(df_sber['name']==using_name)&(df_sber['region']==using_region)&(date_start <= df_sber['date']) & (df_sber['date']<= date_end)]
```

```python
df_sber_print['date'] = pd.to_datetime(df_sber_print['date'])
```

```python
plt.plot(df_sber_print['date'], df_sber_print['value'], color = 'green')
rcParams["figure.figsize"] = 16, 4
plt.title(f"График - {using_name}, по региону: {using_region}")
#plt.xticks(df_sber_print['date'].dt.year)
plt.grid()
plt.show()
```
